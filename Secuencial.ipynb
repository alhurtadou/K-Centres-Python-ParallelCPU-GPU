```python
##IMPLEMENTACIÓN SECUENCIAL

from google.colab import drive
import pandas as pd
import numpy
drive.mount('/content/drive')
#Lectura del csv con pandas
yataros = pd.read_csv('/content/drive/My Drive/Colab Notebooks/yataros1.csv',header=None)

dataset_sec_rec=yataros.copy()
dataset_sec_rec=dataset_sec_rec.transpose()
dataset_sec_rec.drop(0, axis=1, inplace = True)
dataset_sec_rec = dataset_sec_rec.iloc[1:]
dataset_sec_rec = dataset_sec_rec.reset_index(drop=True)
dataset_sec_rec = dataset_sec_rec.to_numpy()
dataset_sec_rec=dataset_sec_rec.astype(numpy.float64)

import numpy as np
from timeit import default_timer
import pandas as pd
import matplotlib.pyplot as plt

def kcentres(d,k=1,nrep=1,elapsed_time=0):
    cents_graph_sec_rec=[]
    m,m2=d.shape

    if(k==1):
        dmax=np.max(d,axis=0)

        dm, jopt=np.min(dmax),np.argmin(dmax)
        labels=np.ones((m,1))
        return labels,jopt,dm

    dmax=np.max(np.max(d,axis=0))
    dopt=np.inf

    for tri in range(1,nrep+1):
        M=np.array([3, 14, 35, 24,39], dtype=np.int64)
        # M=np.random.permutation(m)
        # M=M[0:k]

        J=np.array([-1]*k)

        inicio=default_timer()

        while True:
            dm,I= np.min(d[M,:],axis=0),np.argmin(d[M,:],axis=0)
            # print("dm: ", dm)
            # print("I : ", I )
            for i in range(0,k):
                JJ=np.where(I==i)[0]
                if(len(JJ)==0):
                    J[i]=0
                else:
                    dummy,j,dm=kcentres(d[JJ][:,JJ],1,1)
                    J[i]=JJ[j]
            J=J[J!=-1]
            k=len(J)

            if(len(M)==k and np.all(M==J)):
                #Convergen.
                dmin,labs=np.min(d[J,:],axis=0),np.argmin(d[J,:],axis=0)
                dmin=np.max(dmin)
                cents_graph_sec_rec.append(J.tolist())
                break

            # print("se va a hacer el intercambio", M , " - ", J)
            # M=J
            M = J.copy()

        fin=default_timer()
        elapsed_time+=(fin-inicio)

        #Almacena los centros optimos en JOPT.
        # print("Esto es dmin ",dmin, " esto es ", dopt)
        if(dmin<=dopt):
            dopt=dmin
            labels=np.transpose(labs)
            jopt=J.tolist()


    # Determina los mejores centros de los N ensayos
    dm=np.zeros(k)
    for i in range(k):
        L=np.where(labels==i)[0]
        dm[i]=np.max(d[jopt[i],L])

    return labels,jopt,dm,elapsed_time

def euclidean_distances(data_set):
    num_objects = len(data_set)
    distances = np.zeros((num_objects, num_objects))

    print("-",data_set.dtype)

    for i in range(num_objects):
        for j in range(i+1, num_objects):  # Calcular solo la mitad superior de la matriz
            distances[i, j] = np.linalg.norm(data_set[i] - data_set[j])  # Cálculo de la distancia euclidiana
            distances[j, i] = distances[i, j]  # Rellenar la mitad inferior de la matriz (simétrica)

    return distances



elapsed_time_dis=0
inicio=default_timer()
disi_mat=euclidean_distances(dataset_sec_rec)

fin=default_timer()
elapsed_time_dis=(fin-inicio)
print("Tiempo de Matriz de distancias: ",elapsed_time_dis)

elapsed_time=0
labels,jopt,dm,elapsed_time=kcentres(disi_mat,5,25,elapsed_time)
print("\nTiempo de K-centres: ",elapsed_time/25)



print("\nK.CENTRES secuencial Rec")
print(labels,dm)

print("\nCENTROIDES: ",jopt)
```
